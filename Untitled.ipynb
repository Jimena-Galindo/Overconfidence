```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib as plt
import seaborn as sns
from pathlib import Path
```

```python
others = pd.read_csv(Path.cwd().joinpath('demo_data.csv'))
```

```python
for c in others.columns:
    print(c)
```

```python
others = others.loc[(others['participant.verbal_score'].notnull()) &
                    (others['participant.math_score'].notnull()) &
                    (others['participant.science_score'].notnull()) &
                    (others['participant.sports_score'].notnull()) &
                    (others['participant.us_score'].notnull()) &
                    (others['participant.pop_score'].notnull()) &
                    (others['participant.gender'].notnull()) &
                    (others['participant.nationality'].notnull()) &
                    (others['participant.major'].notnull())
                    , ['participant.code', 
               'participant.verbal_score', 
               'participant.math_score', 
               'participant.us_score', 
               'participant.science_score',
               'participant.pop_score',
               'participant.sports_score', 
               'participant.gender',
               'participant.major',
               'participant.nationality']]

others.reset_index(inplace=True)
```

```python
path_others = Path('Clean/others.csv')  
path_others.parent.mkdir(parents=True, exist_ok=True)  
others.to_csv(path_others)
```
